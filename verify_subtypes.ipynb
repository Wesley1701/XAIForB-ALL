```python
import pandas as pd
import pyarrow.parquet as pq
```

```python
excell_df = pd.read_excel('./assets/subtype_sheet.xlsx', sheet_name='RNA-Seq 1148')

excell_df.head()

excell_sample_ids = excell_df['Sample ID'].tolist()

print(excell_sample_ids)
```

```python
df = pq.read_table('./data/dataset.pq').to_pandas()

df_columns = df.columns.tolist()

print(df_columns)
```

```python
common_sample_ids = set(excell_sample_ids).intersection(set(df_columns))

print(f'Number of common sample IDs: {len(common_sample_ids)}')
```
